```python
cube_duration = 3600 # s
cube_samp = 0.3 # s
lat = -24.59
dec = -5

# hour angles
ha = cube_duration/3600/24*360
# angles in rad
ncube = cube_duration/cube_samp
hr = np.deg2rad(np.linspace(-ha/2, ha/2, ncube))
dr = np.deg2rad(dec)
lr = np.deg2rad(lat)
# parallactic angle in deg
pa = -np.rad2deg(np.arctan2(-np.sin(hr), np.cos(dr)*np.tan(lr) \
        - np.sin(dr)*np.cos(hr)))

plot(pa)
```

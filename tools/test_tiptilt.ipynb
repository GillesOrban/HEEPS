```python
import heeps
from heeps.wavefront.get_array import show_wf
from heeps.util.coord import mas2rms
import proper
import matplotlib.pyplot as plt
%matplotlib inline
```

## Initialize wavefront

```python
# tiptilt in mas
tt_mas = [20, 0]
# config params
conf = dict(
    file_pupil = '',
    diam_ext = 36.905,
    diam_int = 0,
    spi_width = 0,
    seg_width = 0,
)
conf = heeps.config.read_config(verbose=False, **conf)
```

## Test tiptilt with pupil_img_size = diam_nominal

```python
conf['pupil_img_size'] = conf['diam_nominal']
conf = heeps.config.update_config(saveconf=True, verbose=False, **conf)
wf = heeps.pupil.pupil(savefits=True, verbose=False, **conf)
proper.prop_zernikes(wf, [2,3], mas2rms(tt_mas, conf['diam_nominal']))
pup = show_wf(wf, 'phi', npupil=285); plt.colorbar();
```

```python
wf = heeps.pupil.pupil(savefits=True, verbose=True, **conf)
proper.prop_zernikes(wf, [2,3], mas2rms(tt_mas, conf['pupil_img_size']))
pup = show_wf(wf, 'phi', npupil=285); plt.colorbar();
```

## Test tiptilt with pupil_img_size = 2*diam_nominal

```python
conf['pupil_img_size'] = 2*conf['diam_nominal']
conf = heeps.config.update_config(saveconf=True, verbose=False, **conf)
wf = heeps.pupil.pupil(savefits=True, verbose=False, **conf)
proper.prop_zernikes(wf, [2,3], mas2rms(tt_mas, conf['diam_nominal']))
pup = show_wf(wf, 'phi', npupil=285); plt.colorbar();
```

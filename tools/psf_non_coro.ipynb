```python
import heeps
import numpy as np
import os.path
from astropy.io import fits
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
conf = dict(
    dir_output = 'psf_non_coro',
    mode = 'ELT',
    file_pupil = 'pupil/ELT_fullM1.fits',
    file_phase = 'wavefront/cube_Cbasic_20201130_3600s_300ms_0piston_meters_scao_only_285.fits',
    nframes = 1000, # 5 minutes
)
conf = heeps.config.read_config(**conf)
```

```python
conf.update(band='L', hfov=1.1)
conf = heeps.config.update_config(saveconf=True, verbose=True, **conf)
wf = heeps.pupil.pupil(savefits=True, verbose=True, **conf)
phase_screens, amp_screens, tiptilts, misaligns = heeps.wavefront.load_errors(verbose=True, **conf)
psfs = heeps.wavefront.propagate_cube(wf, phase_screens=phase_screens, \
    amp_screens=amp_screens, tiptilts=tiptilts, misaligns=misaligns, verbose=True, **conf)
ELT_psf_L = np.mean(psfs, 0)
```

```python
fits.writeto(os.path.join(conf['dir_output'], 'ELT_psf_L.fits'), np.float32(ELT_psf_L), overwrite=True)
plt.imshow(np.log(ELT_psf_L), origin=True);
```

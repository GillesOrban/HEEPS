```python
from heeps.util.img_processing import get_radial_profile
from astropy.io import fits
import matplotlib.pyplot as plt
import numpy as np
```

```python
folder = '/Users/cdelacroix/Desktop/znse_artemis'
files = ['20230609_131353.fits',
         '20230609_131357.fits',
         '20230609_131402.fits',
         '20230609_131435.fits',
         '20230609_131439.fits',
         '20230609_131444.fits',
         '20230609_131504.fits',
         '20230609_131509.fits',
         '20230609_131513.fits']
xoyo = (75, 75)
xmax = 50
part = '003'
```

```python
y = np.empty((len(files),xmax))
data = np.empty((len(files),150,150))
for i,file in enumerate(files):
    data[i] = fits.getdata(folder + '/' + file)
    y[i] = get_radial_profile(data[i], xoyo, nbin=1)[:xmax]
    if i+1 in [1,2,3,4,5,6,7,8,9]:#[3,5,7]:
        plt.plot(y[i], label=i+1)
plt.legend();
```

```python
plt.plot(y[3]/y[0]);
plt.savefig('radial_trans_%s.png'%part)
```

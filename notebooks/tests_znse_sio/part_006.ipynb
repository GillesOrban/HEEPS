```python
from heeps.util.img_processing import get_radial_profile, fit_gauss_2D
from astropy.io import fits
import matplotlib.pyplot as plt
import numpy as np
```

# L band

```python
folder = '/Users/cdelacroix/INSTRUMENTS/METIS/znse_artemis/L2'
files = ['20230609_133015.fits',
         '20230609_133020.fits',
         '20230609_133024.fits',
         '20230609_133053.fits',
         '20230609_133057.fits',
         '20230609_133102.fits',
         '20230609_133121.fits',
         '20230609_133125.fits',
         '20230609_133130.fits']
xoyo = (75, 75)
xmax = 50
part = '006'
```

```python
y = np.empty((len(files),xmax))
data = np.empty((len(files),150,150))
for i,file in enumerate(files):
    data[i] = fits.getdata(folder + '/' + file)
    y[i] = get_radial_profile(data[i], xoyo, nbin=1)[:xmax]
    if i+1 in [1,2,3,4,5,6,7,8,9]:#[3,5,7]:
        plt.plot(y[i], label=i+1)
plt.legend();
print(np.median(y[3])/np.median(y[0]))
```

```python
plt.plot(y[3]/y[0]);
plt.savefig('radial_trans_%s.png'%part)
print(np.median(y[3]/y[0]))
```

# K band

```python
folder = '/Users/cdelacroix/INSTRUMENTS/METIS/znse_artemis/K'
files = ['20230612_152745.fits',
         '20230612_152750.fits',
         '20230612_152754.fits',
         '20230612_152815.fits',
         '20230612_152820.fits',
         '20230612_152824.fits',
         '20230612_152842.fits',
         '20230612_152846.fits',
         '20230612_152850.fits']
xoyo = (75, 75)
xmax = 50
part = '006'
```

```python
y = np.empty((len(files),xmax))
data = np.empty((len(files),150,150))
for i,file in enumerate(files):
    data[i] = fits.getdata(folder + '/' + file)
    xoyo = fit_gauss_2D(data[i])[1:3]
    xoyo = (int((round(xoyo[0]))), int((round(xoyo[1]))))
    #print(xoyo)
    y[i] = get_radial_profile(data[i], xoyo, nbin=1)[:xmax]
    if i+1 in [1,2,3,4,5,6,7,8,9]:#[1,4]:
        plt.plot(y[i], label=i+1)
plt.legend();
print(np.sum(y[3])/np.sum(y[0]))
plt.xscale('log')
```

```python
import heeps
import matplotlib.pyplot as plt
import numpy as np
import proper
from copy import deepcopy
```

```python
ngrid = 255
wf = heeps.pupil.pupil(ngrid=ngrid, npupil=ngrid, norm_I=True, 
                        pupil_img_size=1, diam_ext=1, diam_int=0, spi_width=0)
pup = heeps.wavefront.get_array.get_wf(wf, 'amp')
plt.imshow(pup, origin='lower');
print('A=%.2f, I=%.2f'%(np.sum(pup), np.sum(pup**2)))
```

```python
mask = np.ones(pup.shape)
center = ngrid // 2
mask[:center] = 0
wf1 = deepcopy(wf)
proper.prop_multiply(wf1, mask)
pup1 = heeps.wavefront.get_array.get_wf(wf1, 'amp')
plt.imshow(pup1, origin='lower');plt.colorbar();
Amax = np.max(pup1)
print('A=%.2f, I=%.2f'%(np.sum(pup1), np.sum(pup1**2)))
```

```python
mask = np.ones(pup.shape)*0.5
wf2 = deepcopy(wf)
proper.prop_multiply(wf2, mask)
pup2 = heeps.wavefront.get_array.get_wf(wf2, 'amp')
plt.imshow(pup2, origin='lower', vmax=Amax);plt.colorbar();
print('A=%.2f, I=%.2f'%(np.sum(pup2), np.sum(pup2**2)))
```

```python
from heeps.pupil import create_stop
import matplotlib.pyplot as plt
from astropy.io import fits
```

```python
conf = dict(    
    seg_width=0,
    pupil_img_size=1, 
    diam_nominal=1, 
    npupil=1001, 
    spi_width=0,
)
```

Margins are calculated wrt nominal diameter, and applied (added/subtracted) 
to the external/internal diameters, and spider width.

```python
pup = create_stop(d_ext=0.8, d_int=0.2, dRext=0, dRint=0, dRspi=0, **conf)
plt.imshow(pup, origin='lower');
fits.writeto('ls_ext_int.fits', pup, overwrite=True)
```

```python
pup = create_stop(d_ext=0.8, d_int=0.2, dRext=0.2, dRint=.2, dRspi=0, **conf)
plt.imshow(pup, origin='lower');
fits.writeto('ls_drext_drint.fits', pup, overwrite=True)
```

```python
conf['spi_width'] = .1
pup = create_stop(d_ext=0.8, d_int=0.2, dRext=0, dRint=0, dRspi=0, **conf)
plt.imshow(pup, origin='lower');
fits.writeto('ls_spi.fits', pup, overwrite=True)
```

```python
pup = create_stop(d_ext=0.8, d_int=0.2, dRext=0, dRint=0, dRspi=.1, **conf)
plt.imshow(pup, origin='lower');
fits.writeto('ls_drspi.fits', pup, overwrite=True)
```

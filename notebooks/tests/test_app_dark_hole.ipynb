```python
import heeps
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

# 1. IMG
### SCAO only, long exposure (100 frames)

```python
conf = dict(
    mode = 'APP',
    f_phase = 'wavefront/cfull/cube_Cfull_20220512_3600s_300ms_0piston_meters_scao_only_L_285.fits',
    f_app_amp = 'optics/METIS_IMG_L_amp.fits',
    f_app_phase = 'optics/METIS_IMG_L_phase.fits',
    add_phase = True,
    nframes_avg = 100,
    hfov = 2.13,
)
conf = heeps.config.read_config(**conf)
verbose = True
conf = heeps.config.update_config(**conf, verbose=verbose)
wf = heeps.pupil.pupil(**conf, verbose=verbose)
psf = heeps.wavefront.propagate(wf, avg=True, **conf, verbose=verbose)
plt.imshow(np.log10(psf), origin='lower');
```

### diffraction limited

```python
conf.update(add_phase=False, nframes_avg=1)
verbose = False
wf = heeps.pupil.pupil(**conf, verbose=verbose)
psf = heeps.wavefront.propagate(wf, avg=True, **conf, verbose=verbose)
plt.imshow(np.log10(psf), origin='lower');
```
